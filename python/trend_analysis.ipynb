```python
import pandas as pd
import numpy as np
```

```python
preds_sg_d=pd.read_csv('../data/sg_d.csv')
preds_p_d=pd.read_csv('../data/p_d.csv')
pred_length=preds_sg_d.shape[0]
weeks=preds_sg_d.shape[1]
preds_sg_d
```

```python
def get_all_rates(sg_d,p_d):
    rates=pd.DataFrame()
    rates["d_sg"]=1/sg_d 
    rates["sg_d"]=sg_d
    rates["d_p"]=1/p_d 
    rates["p_d"]=p_d
    rates["sg_p"]=sg_d/p_d 
    rates["p_sg"]=p_d/sg_d 
    return rates
```

```python
def get_winning_currency(rates):
    difference=np.empty(6)
    for i in range(0,6):
        difference[i]=rates.iloc[pred_length-1,i]-rates.iloc[0,i]
    index=np.argmax(difference)
    print(difference[index],"\t",index+1)
```

```python
for i in range(0,weeks):
    rates=get_all_rates(preds_sg_d.iloc[0:pred_length,i].values,\
                  preds_p_d.iloc[0:pred_length,i].values)
    get_winning_currency(rates)
    #print(rates)
    
```
